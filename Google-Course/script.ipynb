```python
def get_event_date(event):
    return event.date

def current_users(events):
    events.sort(key = get_event_date)
    machines={}
    for event in events:
        if event.machine not in machines:
            machines[event.machine] = set()
        if event.type == 'login':
            machines[event.machine].add(event.user)
        elif event.type == 'logout':
            machines[event.machine].remove(event.user)
        return machines
    
def report_generate(machines):
    for machine, users in machines.items():
        if len(users) > 0:
            user_list = ", ".join(users)
            print("{}: {}".format(machine,user_list))
```

```python
class Event():
    def __init__(self,event_date,event_type,machine_name,user):
        self.date = event_date
        self.type = event_type
        self.machine = machine_name
        self.user = user
```

```python
events = [
    Event('2023-02-20 08:44:30', "login", "myworkstation.local", 'Jordan'),
    Event('2023-02-20 08:44:45', 'login', 'Webserver.local', 'Dhaka'),
    Event('2023-02-20 08:45:15', 'logout', 'webserver.out', 'amsterdam'),
]
```

```python
users = current_users(events)
print(users)
```

```python
report_generate(users)
```

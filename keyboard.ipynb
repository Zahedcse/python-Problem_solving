```python
import tkinter as tk

def key_press(event):
    if event.keysym == "BackSpace":
        current_text = label.cget("text")
        label.config(text=current_text[:-1])
    elif event.keysym == "Return":
        label.config(text=label.cget("text") + "\n")
    else:
        label.config(text=label.cget("text") + event.char)

root = tk.Tk()
frame = tk.Frame(root, width=500, height=500)
frame.pack()

keys = [
    ['`', '1', '2', '3', '4', '5', '6', '7', '8', '9', '0', '-', '=', 'Backspace'],
    ['Tab', 'q', 'w', 'e', 'r', 't', 'y', 'u', 'i', 'o', 'p', '[', ']', '\\'],
    ['Caps Lock', 'a', 's', 'd', 'f', 'g', 'h', 'j', 'k', 'l', ';', '\'', 'Enter'],
    ['Shift', 'z', 'x', 'c', 'v', 'b', 'n', 'm', ',', '.', '/', 'Shift'],
    ['Ctrl', 'Alt', 'Space', 'Alt', 'Ctrl']
]

for row in range(len(keys)):
    for col in range(len(keys[row])):
        button = tk.Button(frame, text=keys[row][col], height=2, width=10)
        button.grid(row=row, column=col)
        button.bind("<Button-1>", lambda event, btn=keys[row][col]: click(btn))

def click(key):
    if key == "Backspace":
        current_text = label.cget("text")
        label.config(text=current_text[:-1])
    elif key == "Enter":
        label.config(text=label.cget("text") + "\n")
    elif key == "Space":
        label.config(text=label.cget("text") + " ")
    elif key == "Caps Lock":
        pass # Add logic for handling caps lock here
    else:
        label.config(text=label.cget("text") + key)

label = tk.Label(frame, text="", font=("Arial", 14), wraplength=450, justify=tk.LEFT)
label.grid(row=len(keys), column=0, columnspan=len(keys[0]), sticky='W')

frame.bind("<Key>", key_press)
frame.focus_set()

root.mainloop()
```
